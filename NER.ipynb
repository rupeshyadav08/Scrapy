```python
import wikipedia
```

```python
data=(wikipedia.summary("America"))
```

```python
data
```

```python
#import required library
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
```

```python
#apply word tokenization and part-of-speech tagging

def preprocess(sentence):
    sentence = nltk.word_tokenize(sentence)
    sentence = nltk.pos_tag(sentence)
    return sentence
```

```python
sentence = preprocess(data)
sentence
```

```python
#noun phrase chunking to identify named entities using a regular expression 
pattern = 'NP: {<DT>?<JJ>*<NN>}'
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sentence)
print(cs)

```

```python
#IOB 
#to represent chunk structures in files, and we will also be using this format.

from nltk.chunk import conlltags2tree, tree2conlltags
from pprint import pprint
iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

```python
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
doc=nlp(data)
pprint([(X.text, X.label_) for X in doc.ents])
```

```python
pprint([(X, X.ent_iob_, X.ent_type_) for X in doc])
```

```python
ny_bb=data
article = nlp(ny_bb)
len(article.ents)
```

```python
#Representation of most common articles

labels = [x.label_ for x in article.ents]
Counter(labels)
```

```python
#Excration of most common word

items = [x.text for x in article.ents]
Counter(items).most_common(9)
```
